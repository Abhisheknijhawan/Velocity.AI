# Functions To Calculate Different Angles

```python
# Function to calculate hip joint angle
def calculate_hip_joint_angle(knee,hip,ankle):
    angle = math.degrees(math.atan2(knee[1] - hip[1], knee[0] - hip[0]) - math.atan2(ankle[1] - knee[1], ankle[0] - knee[0]))
    return angle
```

```python
cap = cv2.VideoCapture(input('Enter video file name ')+'.mp4') # To use live camera write cv2.VideoCapture(0)
numberOfSteps= 0

frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))



step_lengths = []  #Array to store the values of step Lengths Done

Stride_Lengths = [] #Arrays to store the value of Stride Lengths Done

hip_joint_angle_extension  = [] # Arrays to store the values of hip joint angle extension

hip_joint_angle_flexion = [] # Arrays to store the vlaues of hip joint angle flexion

knee_joint_angle = []  # Arrays to store the vlaues of knee joint angle

step_time = [] # Array to store the values of step time

stance_time = [] #Array to store the values of stance time

swing_time = [] #Array to store the values of swing time

gait_speed = [1] #Array to store the values of gait speed



# variables used for gait speed
prev_time = 0
prev_left_foot_pos = None
gait_speed = None

## Setup mediapipe instance

with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
    
    #while cap.isOpened(): #(Commneted this because this is used when we need live camera)

    for i in range(frame_count): # when we are using live camera comment this line
        
        ret, frame = cap.read()
        
        # Recolor image to RGB
        
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        
        image.flags.writeable = False
      
        # Make detection
        
        results = pose.process(image)
    
        # Recolor back to BGR
        
        image.flags.writeable = True
        
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        
        # Extract landmarks
        
        try:
            
            landmarks = results.pose_landmarks.landmark
            
            # All required coordinates
        
            
            Left_hip = [landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].x,landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].y]
            
            right_hip = [landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].x,landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].y]
            
            Left_knee = [landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].x,landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].y]
            
            right_knee = [landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].x,landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].y]
            
            Left_ankle = [landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].x,landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].y]
            
            right_ankle = [landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].x,landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].y]
            
            Left_heel = [landmarks[mp_pose.PoseLandmark.LEFT_HEEL.value].x,landmarks[mp_pose.PoseLandmark.LEFT_HEEL.value].y]
            
            right_heel = [landmarks[mp_pose.PoseLandmark.RIGHT_HEEL.value].x,landmarks[mp_pose.PoseLandmark.RIGHT_HEEL.value].y]
            
            left_foot_index= [landmarks[mp_pose.PoseLandmark.LEFT_FOOT_INDEX.value].x,landmarks[mp_pose.PoseLandmark.LEFT_FOOT_INDEX.value].y]
            
            right_foot_index= [landmarks[mp_pose.PoseLandmark.RIGHT_FOOT_INDEX.value].x,landmarks[mp_pose.PoseLandmark.RIGHT_FOOT_INDEX.value].y]
            
                 
            # 1 Calculate hip joint angle (extension)
            # 2 Calculate hip joint angle (flexion)
            # 3 Calculate knee joint angle
            # 4 Calculate hip-knee ankle angle
            # 5 Step Time
            # 6 Stance time
            # 7 Swing Time
            # 8 Step length done
            # 9 step width  done
            # 10 Gait Speed

            # 11 Stride length  done
            

        
            
            # calculate distance 
            
            
            
            step_length = (math.sqrt((Left_heel[1] - Left_heel[0]) ** 2 + (right_heel[1] - right_heel[0]) ** 2))*100
            #step length is basically a distance between two points we use distance formulla to calculate it in cm thats
            # we multiple it with 100
            numberOfSteps+=1
            
            
            step_lengths.append(step_length)
            
            Stride_Lengths.append(step_length*2)
             
            
        except:
            
            pass
        
        
        
        # Render detections
        
        mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS,
                                  
                                mp_drawing.DrawingSpec(color=(245,117,66), thickness=2, circle_radius=2), 
                                  
                                mp_drawing.DrawingSpec(color=(245,66,230), thickness=2, circle_radius=2) 
                                  
                                 )               
        
        cv2.imshow('Kunal Anshuman Lakshay Human Gait Code', image)

        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

            
    # Calculate the mean values of the parameters
    
    step_length_mean = np.mean(step_lengths)
    

    
    stride_length_mean = step_length_mean*2
    
    
    # Printing the mean values of parameters
    
    print("Step Length in cm is : ",step_length_mean , "cm")
    
    print("Stride Length in cm  is : ",stride_length_mean , "cm")
    
    print("The number of steps in this video is :", numberOfSteps)
    
    
    cap.release()
    
    cv2.destroyAllWindows()
```
